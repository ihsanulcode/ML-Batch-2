# Understanding Data Visualization

Data visualization is important for interpreting complex information quickly and efficiently. It helps uncover patterns, trends, and relationships within data that might otherwise be difficult to discern. There are various types of data visualization, including charts, graphs, maps, and infographics. Each type is suited to different data sets and analytical purposes, allowing for clearer communication of insights.


# Types of Data Visualization

There are various types of data visualization techniques, each suitable for different data types and intended insights. For examples:

* **[Bar Charts](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.bar.html):** Ideal for comparing categorical data by showing the relationships between distinct categories.

* **[Line Charts](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.plot.html):** Display trends over time or correlations between variables, commonly used in tracking changes.

* **[Pie Charts](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.pie.html):** Show parts of a whole, useful for illustrating proportions or percentages.

* **[Scatter Plots](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.scatter.html):** Demonstrate relationships between two variables, showcasing correlations or clusters.

* **[Histograms](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.hist.html):** Visualize distributions of numerical data by showing the frequency of values within specified ranges.

* **[Heatmaps](https://seaborn.pydata.org/generated/seaborn.heatmap.html):** Depict data values in a matrix format using colors to represent variations in the dataset.

* **[Box Plots](https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.boxplot.html):** Display the distribution of numerical data through quartiles, identifying outliers and variability. [seaborn.boxplot](https://seaborn.pydata.org/generated/seaborn.boxplot.html)

And many more...

# Data Visualization using Matplotlib & Seaborn

Matplotlib and Seaborn are often used together because they complement each other in the realm of data visualization:

1. **Matplotlib** is a foundational library that offers extensive flexibility in creating a wide range of plots and graphs. It allows for precise customization of visuals, making it powerful but sometimes requiring more lines of code for intricate designs.

2. **Seaborn** is built on top of Matplotlib and offers a higher-level interface for creating attractive statistical graphics. It simplifies the creation of complex visuals by providing default styles and themes. Seaborn also specializes in statistical plots that showcase relationships in data, such as regression plots, violin plots, and correlation matrices.

Combining the two libraries allows you to leverage the strengths of both. Matplotlib's flexibility can be utilized for detailed customization when needed, while Seaborn's concise syntax and built-in statistical functionalities make it easy to create visually appealing and informative plots quickly.

Moreover, Seaborn can directly use Matplotlib functions, making it seamless to integrate the two. For instance, you can create a plot using Seaborn and then further customize it using Matplotlib's extensive capabilities. This synergy between the libraries enables data scientists and analysts to create polished and insightful visualizations efficiently.

# Bar Chart

`Matplotlib Bar chart documentation`
https://matplotlib.org/3.5.3/api/_as_gen/matplotlib.pyplot.bar.html

```python
import matplotlib.pyplot as plt
import pandas as pd

# Creating a sample dataframe
data = {
    'Categories' : ['Category A', 'Category B', 'Category C', 'Category D'],
    'Values' : [25, 40, 30, 35]
}
df = pd.DataFrame(data)

# Plotting the bar chart using matplotlib bar function
plt.bar(df['Categories'], df['Values'])
# Adding lables and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Fig: Category vs Values')
# Add grid
plt.grid()
# Display the chart
plt.show()
```

```python
# Using Seaborn
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Creating a sample dataframe
data = {
    'Categories' : ['Category A', 'Category B', 'Category C', 'Category D'],
    'Values' : [25, 40, 30, 35]
}
df = pd.DataFrame(data)

# Plotting the bar plot using sns
sns.barplot(x=df.Categories, y=df.Values)
# Adding lables and title
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Fig: Category vs Values')

# Display the chart
plt.show()
```

# Choosing Seaborn Color Palettes
`Seaborn` https://seaborn.pydata.org/tutorial/color_palettes.html

`Matplotlib` https://matplotlib.org/stable/gallery/color/named_colors.html

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample df
data = {
    'Categories': ['A','B','C','D'],
    'Values': [25,40,30,35]
}
df = pd.DataFrame(data)

# list of color palettes
palette_list = ['deep', 'muted', 'bright', 'pastel', 'dark', 'colorblind']
# Set figure sie
plt.figure(figsize=(12,8))

for i, palette in enumerate(palette_list):
    plt.subplot(2,3, i+1)
    sns.barplot(data=df, x=df.Categories, y=df.Values, palette=palette)
    plt.title(f'Palette: {palette}')

# Applying tight layout
plt.tight_layout()
# Show the plot
plt.show()
```

# Function to Save Figures

```python
# code to save the figures as high-res PNGs
import os
from pathlib import Path
from matplotlib import pyplot as plt

IMAGES_PATH = Path() / "Images/Data_Visualization"
IMAGES_PATH.mkdir(parents=True, exist_ok=True)

def save_fig(fig_id, tight_layout=True, fig_extension="png", resolution=300):
    path = IMAGES_PATH / f"{fig_id}.{fig_extension}"
    if tight_layout:
        plt.tight_layout()
    plt.savefig(path, format=fig_extension, dpi=resolution)
```

# Pie Chart

`Matplotlib Pie Chart documentation` https://matplotlib.org/stable/gallery/pie_and_polar_charts/pie_features.html

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Sample df
data = {
    'Categories': ['A','B','C','D'],
    'Values': [25,40,30,35]
}
df = pd.DataFrame(data)

# Set fig size
plt.figure(figsize=(6,6))
# Create a pie chart
plt.pie(df['Values'], 
        labels=df['Categories'], 
        autopct='%1.1f%%', 
        startangle=90, 
        colors=['salmon', 'orange', 'khaki', 'lightcyan'])

plt.title('Pie Chart')
# Save the figure
save_fig('Matplotlib Pie Chart')
# Show plot
plt.show()
```

# Count Plot


A count plot, in the context of data visualization using libraries like Seaborn or Matplotlib, is a type of categorical plot that displays the count of observations in each category using bars.

The count plot is particularly useful when dealing with categorical data, where you want to visualize the frequency or distribution of observations within different categories. It represents the number of occurrences of each category in the dataset using vertical bars.

https://seaborn.pydata.org/generated/seaborn.countplot.html

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Create a custom DataFrame
data = {
    'Stroke' : ['Yes', 'No', "Not Sure", "Yes", "No", "Not Sure", 'Yes', 'No','Maybe', 'Yes']
}
df = pd.DataFrame(data)

# Creating a count plot using sns
plt.figure(figsize=(8,6))
sns.countplot(data=df, x=df.Stroke)
plt.title('Stroke Count')
plt.xlabel("Stroke Status")
plt.ylabel("Count")

# use grid
plt.grid()
# Save the figure
save_fig('Stroke Value Count Plot')
# Show the fig
plt.show()
```

# Hist Plot

A histplot, also known as a histogram, is a type of plot used to represent the distribution of a single numeric variable.
https://seaborn.pydata.org/generated/seaborn.histplot.html

```python
import pandas as pd
# Import dataset
df = pd.read_csv('Datasets/healthcare-dataset-stroke-data.csv')
df.info()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a histogram or histplot using sns
sns.histplot(data=df, x=df.age, color='y', kde=True)

# Add labels and title
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.title('Age Distribution')
# plt.grid()
# Save the figure
save_fig('Age Distribution')

# Show the plot
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Create a histogram or histplot using sns
sns.histplot(data=df, x=df.bmi, color='b', kde=True)

# Add labels and title
plt.xlabel('BMI')
plt.ylabel('Frequency')
plt.title('BMI Distribution')
# plt.grid()
# Save the figure
save_fig('BMi Distribution')

# Show the plot
plt.show()
```

# Difference Between Count and Hist Plot

The main differences between a count plot and a histplot lie in the types of data they are used to visualize.

**Key Differences:**
**Data Type:** Count plots deal with categorical data and show the count of occurrences within each category. Histplots handle numeric data, showing the distribution and frequency of values.

**Variable Representation**: Count plots display categorical variables as bars with the count on the y-axis. Histplots show the distribution of numeric values as bars representing bins on the x-axis.

**Usage:** Count plots are suitable for categorical variables' frequency comparison, while histplots are used to understand the distribution of continuous variables.

# Swarm Plot

A swarm plot is a categorical scatter plot that displays individual data points along an axis based on categories. It's especially useful for visualizing the distribution of categorical data points and their underlying distributions within each category.

**Key Features of Swarm Plot:**

* **Categorical Representation:** It represents individual data points as points along the categorical axis, avoiding overlapping by positioning points in a way that minimizes collision.

* **Data Distribution:** Swarm plots help in visualizing the distribution of data points within each category or group.

* **Point Density:** Points within a category are spread out along the axis, providing a sense of their density and distribution.

**Characteristics and Usage:**

* **Category Comparison:** Swarm plots allow easy comparison of data distributions across different categories.

* **Presence of Outliers:** They help identify outliers and irregularities within each category by displaying individual data points.
Data Point Distribution: Each point represents an individual observation, providing a clearer view of the distribution compared to traditional scatter plots.

**Interpretation:**

* **Point Distribution:** The horizontal position of points within each category represents their values or properties, while the vertical spacing helps avoid overlap and shows the density of observations.

* **Patterns and Clusters:** Swarm plots can reveal patterns, clusters, or gaps in the data distribution within each category.

**Limitations:**

* **Space Requirement:** As data points are spread out to avoid overlap, swarm plots might not be suitable for very large datasets or categories with numerous data points.

* **Potential Overplotting:** In cases of a high density of data points, overlapping might occur, reducing readability.

https://seaborn.pydata.org/generated/seaborn.swarmplot.html

# Box Plot

A box plot, also known as a box-and-whisker plot, is a graphical representation that displays the distribution and statistical summary of a continuous variable through five key summary statistics. It provides a concise and effective way to understand the distribution of data and identify potential outliers.

![image.png](https://media.licdn.com/dms/image/C5612AQHmj-jx1e72iA/article-inline_image-shrink_400_744/0/1592543224278?e=1708560000&v=beta&t=4MIb6fR1bcDD-kPW79lvYVJ6C-skeixKLgSgN2yFy84)

**Key Features of a Box Plot:**

**Five Summary Statistics:** A box plot represents the following statistical measures:

* `Minimum:` The smallest data point within the dataset excluding outliers.
First Quartile (Q1): The value below which 25% of the data falls (25th percentile).

* `Median (Q2):` The middle value of the dataset (50th percentile or second quartile).

* `Third Quartile (Q3):` The value below which 75% of the data falls (75th percentile).

* `Maximum:` The largest data point within the dataset excluding outliers.

![image.png](https://sphweb.bumc.bu.edu/otlt/mph-modules/bs/bs704_summarizingdata/Interquartile-Odd.png)

**Characteristics and Usage:**

* **Central Tendency:** Box plots provide insights into the central tendency and spread of the data.

* **Outlier Identification:** They help identify potential outliers in the dataset.

* **Comparison:** Box plots facilitate comparisons between different datasets or categories.

**Interpretation:**

* **Box:** The box in the plot represents the interquartile range (IQR) where the middle 50% of the data lies (Q1 to Q3).

* **Whiskers:** The whiskers extend to the minimum and maximum values within a calculated range, excluding outliers.

* **Outliers:** Individual data points beyond the whiskers are considered outliers.

https://seaborn.pydata.org/generated/seaborn.boxplot.html

# Pair Plot

A pair plot, also known as a pairs plot, is a powerful visualization tool used in exploratory data analysis (EDA) to analyze relationships between multiple variables in a dataset. It plots pairwise relationships between numerical variables, showcasing both distributions of individual variables and correlations between pairs of variables.

https://seaborn.pydata.org/generated/seaborn.pairplot.html
