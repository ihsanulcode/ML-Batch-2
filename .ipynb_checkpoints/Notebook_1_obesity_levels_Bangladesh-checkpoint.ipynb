## Removing Outliers

```python
def remove_outliers_iqr(df, attributes, threshold=1.5):
    df_clean = df.copy()
    for attr in attributes:
        Q1 = df_clean[attr].quantile(0.25)
        Q3 = df_clean[attr].quantile(0.75)
        IQR = Q3 - Q1
        lower_bound = Q1 - (threshold * IQR)
        upper_bound = Q3 + (threshold * IQR)
        df_clean = df_clean[(df_clean[attr] > lower_bound) & (df_clean[attr] < upper_bound)]
    return df_clean

# Attributes that needs to clean
attributes = ['age', 'weight', 'height_meters']

# Call the function for removing outliers
df_cleaned = remove_outliers_iqr(df, attributes)

# Old shape
print('Original df shape:',df.shape)
# New Shape
print('New df shape:',df_cleaned.shape)

# Replace the df
df = df_cleaned
```

```python
# See outliers again
# Set the plot fig size
plt.figure(figsize=(12,8))

# Define the default font sizes
plt.rc('font', size=14)
plt.rc('axes', labelsize=14, titlesize=14)
plt.rc('legend', fontsize=14)
plt.rc('xtick', labelsize=10)
plt.rc('ytick', labelsize=10)

# See the outliers for age attribute
plt.subplot(2, 2, 1)
sns.boxplot(x=df.age)
plt.xlabel("Age")

# See the outliers for weight attribute
plt.subplot(2, 2, 2)
sns.boxplot(x=df.weight)
plt.xlabel("Weight (kg)")

# See the outliers for height attribute
plt.subplot(2, 2, 3)
sns.boxplot(x=df.height_meters)
plt.xlabel("Height (m)")

# Adjust the spacing between subplots
plt.subplots_adjust(left=0.1, right=0.9, bottom=0.1, top=0.9, wspace=0.4, hspace=0.4)

save_fig('removing_outliers')
plt.show()
```

## Ordinal Encoding

```python
# The mapping dictionaries for ordinal encoding
gender_map = {
    'Male': 1,
    'Female': 0
}
boolean_map = {
    'Yes': 1,
    'No': 0
}
eat_vegetables_map = {
    'Never': 0,
    'Sometimes': 1,
    'Always': 2
}
daily_main_meals_map = {
    'Between 1-2': 0,
    'Three': 1,
    'More than Three': 2 
}
food_between_meals_map = {
    'No': 0,
    'Sometimes': 1,
    'Frequently': 2,
    'Always': 3
}
alcohol_map = {
    'I do not drink': 0,
    'Sometimes': 1,
    'Frequently': 2,
    'Always': 3
}
daily_water_intake_map = {
    'Less than a liter': 0,
    'Between 1 and 2 L': 1,
    'More than 2 L': 2
}
physical_exercise_map = {
    'I do not have': 0,
    '1 or 2 days': 1,
    '2 or 4 days': 2,
    '4 or 5 days': 3,
    'Almost Everyday': 4
}
daily_device_usage_map = {
    '0-2 hours': 0,
    '3-5 hours': 1,
    'More than 5 hours': 2
}
transportation_map = {
    'Private Car': 0,
    'Public Transportation': 1,
    'Rickshaw': 2,
    'Bike': 3,
    'Walking' : 4 
}
obesity_level_map = {
    'Insufficient_Weight': 0,
    'Normal_Weight': 1,
    'Overweight': 2,
    'Obesity_Type_I': 3,
    'Obesity_Type_II': 4,
    'Obesity_Type_III': 5
}

# Ordinal encoding
df['gender'] = df['gender'].map(gender_map)
df['family_history_ofOverweight'] = df['family_history_ofOverweight'].map(boolean_map)
df['high_caloric_foodfrequently'] = df['high_caloric_foodfrequently'].map(boolean_map)
df['eat_vegetables'] = df['eat_vegetables'].map(eat_vegetables_map)
df['daily_main_meals'] = df['daily_main_meals'].map(daily_main_meals_map)
df['food_between_meals'] = df['food_between_meals'].map(food_between_meals_map)
df['smoke'] = df['smoke'].map(boolean_map)
df['alcohol'] = df['alcohol'].map(alcohol_map)
df['daily_water_intake'] = df['daily_water_intake'].map(daily_water_intake_map)
df['monitor_calories'] = df['monitor_calories'].map(boolean_map)
df['physical_exercise'] = df['physical_exercise'].map(physical_exercise_map)
df['daily_device_usage'] = df['daily_device_usage'].map(daily_device_usage_map)
df['transportation'] = df['transportation'].map(transportation_map)
df['obesity_level'] = df['obesity_level'].map(obesity_level_map)

# Encoding done
```

#### Experimenting with different values of K

```python
train_acc = []
test_acc = []
neighbors = []

for i in range(1,31):
    neighbors.append(i)
    model = KNeighborsClassifier(n_neighbors = i)
    model.fit(x_train, y_train)
    train_acc.append(accuracy_score(y_train, model.predict(x_train)).round(3))
    test_acc.append(accuracy_score(y_test, model.predict(x_test)).round(3))
    
for i in range(1, 31):
    print('k=',i,', Train Acc. =',train_acc[i-1],', Test Acc. =',test_acc[i-1])

print("Max Train Acc.=",max(train_acc),'for k=',neighbors[train_acc.index(max(train_acc))])
print("Max Test Acc.=",max(test_acc),'for k=',neighbors[test_acc.index(max(test_acc))])

plt.figure(figsize = (8,6))
plt.xlabel('K values')
plt.ylabel('Accuracy')
plt.title('Variation of Accuracy with k')
plt.plot(neighbors, train_acc, color = 'blue', linewidth = 1, label = 'train accuracy');
plt.plot(neighbors, test_acc, color = 'red', linewidth = 2, label = 'test accuracy')
plt.legend()
save_fig('Accuracy against k')
plt.show()
```

#### Analyze the variation of result with Max-Depth

```python
train_acc = []
test_acc = []
max_depth =[]

for depth in range(1, 31):
    clf = DecisionTreeClassifier(max_depth = depth)
    clf.fit(x_train, y_train)
    train_acc.append(accuracy_score(y_train, clf.predict(x_train)).round(3))
    test_acc.append(accuracy_score(y_test, clf.predict(x_test)).round(3))
    max_depth.append(depth)

for i in range(1, 31):
    print('Max Depth=',i,', Train Acc. =',train_acc[i-1],', Test Acc. =',test_acc[i-1])

plt.figure(figsize = (8,6))
plt.xlabel('Max-Depth')
plt.ylabel('Accuracy')
plt.title('Variation of Accuracy with Max-Depth')
plt.plot(max_depth, train_acc, color = 'blue', linewidth = 1, label = 'train accuracy')
plt.plot(max_depth, test_acc, color = 'red', linewidth = 2, label = 'test accuracy')
plt.legend()
save_fig('Accuracy against max_depth')
plt.show()
```
